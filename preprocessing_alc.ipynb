```python
from sklearn.preprocessing import LabelEncoder
```

```python
import numpy as np
```

```python
df=df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
df.head()
```

```python
df.shape
```

```python
# Initialize LabelEncoder
le = LabelEncoder()

# Fit on Label column which contain attack categories
le.fit(df[' Label'])

# Get all classes
classes = list(le.classes_)

# Move benign to index 0
if 'BENIGN' in classes:
    classes.remove('BENIGN')
    classes = ['BENIGN'] + classes  # BENIGN first

# Create a mapping dictionary
label_mapping = {label: idx for idx, label in enumerate(classes)}

# Map attack_cat to numerical labels
df['label'] = df[' Label'].map(label_mapping)

print(label_mapping)
print(df[[' Label', 'label']].head())
```

```python
df.drop(' Label', axis=1, inplace=True)
```

```python
#check for categorical columns
categorical_columns = df.select_dtypes(include=['object', 'category']).columns
print("Categorical columns (by dtype):", categorical_columns.tolist())
```

```python
#obtaining a small fraction of the dataset - 10%
df=df.sample(frac=0.04, random_state=42).reset_index(drop=True)
```

```python
df.to_csv("al.csv",index=False)
```

```python
#convert infinite values into nan
df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
#replace with mean
for col in df.select_dtypes(include=[np.number]).columns:
    df[col]=df[col].fillna(df[col].mean())
```

```python
# Drop duplicates
df = df.drop_duplicates()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Count of each label
label_counts = df['label'].value_counts()
print(label_counts)
```

```python
threshold = 200
```

```python
# Find labels to keep
labels_to_keep = label_counts[label_counts >= threshold].index

# Filter
df_filtered = df[df['label'].isin(labels_to_keep)].reset_index(drop=True)

print("Remaining labels:", df_filtered['label'].value_counts())
```

```python
original_labels = {
    'BENIGN': 0,
    'Bot': 1,
    'DDoS': 2,
    'DoS GoldenEye': 3,
    'DoS Hulk': 4,
    'DoS Slowhttptest': 5,
    'DoS slowloris': 6,
    'FTP-Patator': 7,
    'Heartbleed': 8,
    'Infiltration': 9,
    'PortScan': 10,
    'SSH-Patator': 11,
    'Web Attack – Brute Force': 12,
    'Web Attack – Sql Injection': 13,
    'Web Attack – XSS': 14
}
```

```python

remaining_labels = [0,4,10,2,3,7,6,5,11,1,12,14]

# Sort remaining_labels to keep original order
remaining_labels_sorted = sorted(remaining_labels)  

# Create new mapping: old_label to new_label
old_to_new = {old: new for new, old in enumerate(remaining_labels)}

# Create updated label names mapping: new_label to attack name
new_label_names = {new: [name for name, old in original_labels.items() if old == old_label][0]
                   for old_label, new in old_to_new.items()}

print("Old to New Label Mapping:", old_to_new)
print("New Label Names:", new_label_names)
```

```python
df['label'] = df['label'].map(old_to_new)
```

```python
# Drop rows where label is NaN
df = df.dropna(subset=['label']).reset_index(drop=True)
```

```python
# Save to new CSV
df.to_csv("al_l.csv", index=False)
```

```python
#split the dataset into train and temp
train_df, temp_df = train_test_split(df, test_size=0.3, random_state=42, shuffle=True, stratify=df['label'])
```

```python
#split the train set into test and validation
test_df, val_df = train_test_split(temp_df, test_size=0.5, random_state=42, shuffle=True, stratify=temp_df['label'])
```

```python
# Split into features (X) and label (y)
X = train_df.drop('label', axis=1)
y = train_df['label']
```
